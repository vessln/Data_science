### Problem 1. Read the dataset (1 point)
Read the file without unzipping it first. You can try a different character encoding, like `unicode_escape`. Don't worry too much about weird characters.

First I unzipped the spotify_most_streamed_2024.zip, then I read the df using encoding = "ISO-8859-1". Finally I saved a new .csv file: most_streamed_spotify_songs_2024.csv so I can open the whole table.

```python
zip_path = "data/spotify_most_streamed_2024.zip"
csv_file = "Most Streamed Spotify Songs 2024.csv"

with zipfile.ZipFile(zip_path, "r") as z:
    with z.open(csv_file) as file:
        songs_data = pd.read_csv(file, encoding = "ISO-8859-1")

new_csv_path = "data/most_streamed_spotify_songs_2024.csv"
songs_data.to_csv(new_csv_path, index = False) 
```

### Problem 6. Playlists (2 points)
Is there any connection (correlation) between users adding a song to playlists in one service, or another? Only Spotify, Apple, Deezer, and Amazon offer the ability to add a song to a playlist. Find a way to plot all these relationships at the same time, and analyze them. Experiment with different types of correlations.

```python
songs_data.columns
```

```python
services_playlists_count = ["spotify_playlist_count", "apple_music_playlist_count", "deezer_playlist_count", "amazon_playlist_count"]
```

Pearson correlation coefficient:

```python
songs_data[services_playlists_count].corr()
```

```python
sns.heatmap(songs_data[services_playlists_count].corr(), annot = True) 

plt.title("Pearson correlation between Spotify, Apple, Deezer and Amazon playlist count")
plt.show()
```

Correlation analysis shows that there is strong positive linear relashionship (correlation coefficient = 0.78) between the number of playlists a song is included in Apple Music and Deezer. So the songs popular on Apple Music are probably popular on Deezer. \
Also there is moderate to strong positive linear relashionship (correlation coefficient = 0.69) between the number of playlists a song is included in on Spotify and Apple Music. This suggests that the songs popular on Spotify are probably popular on Apple Music.

Spearman correlation coefficient:

```python
songs_data[services_playlists_count].corr("spearman")
```

```python
sns.heatmap(songs_data[services_playlists_count].corr("spearman"), annot = True, cmap = "YlGnBu") 
plt.title("Pearson correlation between Spotify, Apple, Deezer and Amazon playlist count")
plt.show()
```

This analysis also shows a strong positive monotonic relationship between the playlist counts on Apple Music and Deezer (correlation coefficien = 0.77).

To calculate YouTube views-to-likes ratio I divide youtube views by youtube likes:

```python
songs_data["views_likes_ratio"] = songs_data.youtube_views / songs_data.youtube_likes
mean_ratio = songs_data.views_likes_ratio.mean()
print(f"Mean YouTube views-to-likes ratio is {mean_ratio:.2f}")
```

```python
songs_data.views_likes_ratio.describe()
```

```python
songs_data[songs_data.views_likes_ratio > 1500]
```

If views to likes ratio is too big - there are many views and few likes. For example if views_likes_ratio = 1500 means that for every 1500 views there is 1 like. I think values like these and bigger are outliers so I'm removing them:

```python
songs_data.drop(songs_data[songs_data.views_likes_ratio > 1500].index, inplace = True) 
```

```python
songs_data[songs_data.views_likes_ratio > 1500]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(songs_data.views_likes_ratio, bins = 50, color = "red")

plt.xticks(range(0, 1500, 100))
plt.title("Distribution of mean YouTube views-to-likes ratio")
plt.xlabel("views_likes_ratio")
plt.ylabel("frequency")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.hist(songs_data.youtube_views, bins=50, color = "green")
ax1.set_title("Distribution of Youtube views")
ax1.set_xlabel("views")
ax1.set_ylabel("frequency")

ax2.hist(songs_data.youtube_likes, bins=50, color = "blue")
ax2.set_title("Distribution of Youtube likes")
ax2.set_xlabel("likes")

plt.tight_layout()
plt.show()
```

The visualisations show that the more views a song has, the more likes it has. The mean views_likes ratio is 169 - 1 like for every 169 views.

The most popular songs on TikTok released every year show... interesting behavior. Which years peaked the most TikTok views? Show an appropriate chart. Can you explain this behavior? For a bit of context, TikTok was created in 2016.

Now, how much popular is the most popular song for each release year, than the mean popularity? Analyze the results.

```python
songs_data.groupby(songs_data.release_date.dt.year)["tiktok_views"].apply(lambda x: x.isna().sum())
```

Songs released in 2022 have the most tiktok views. Total views for this year are more than 800 bilions. But this conclusion may not be correct because I don't have enough information about for 2023 and 2024. There are many missing values about tiktok_views column (2023 - 270 missing values,  2024 - 500 missing values).

```python
most_popular_songs = songs_data.loc[songs_data.groupby(songs_data.release_date.dt.year)["tiktok_views"].idxmax()]
```

```python
most_popular_songs.release_date = most_popular_songs.release_date.dt.year
```

The most popular songs for each release year according to tiktok views are:

```python
most_popular_songs[["release_date", "track", "artist", "tiktok_views"]]
```

To find the mean of tiktok views for each year I group by release year and for each year i calculate the mean of the tiktok_views column:

```python
mean_tiktok_views = songs_data.groupby(songs_data.release_date.dt.year)["tiktok_views"].mean()
```

```python
mean_tiktok_views
```

I merge most_popular_songs dataset with the mean_tiktok_views series on the release_date column. Then I compare the most popular songs (the songs with the most tiktok views) for each release year vs. the mean tiktok views for each year.

```python
comparison_mean = most_popular_songs[["release_date", "track", "artist", "tiktok_views"]]
comparison_mean = comparison_mean.merge(mean_tiktok_views, on = "release_date", suffixes = ("_song", "_mean_year"))
```

```python
comparison_mean
```

```python
songs_data.release_date.dt.year.value_counts()
```

```python
comparison_mean["difference"] = comparison_mean["tiktok_views_song"] - comparison_mean["tiktok_views_mean_year"]
comparison_mean.difference
```

```python
comparison_mean
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
comparison_mean.set_index("release_date")[["tiktok_views_song", "tiktok_views_mean_year"]].plot(kind = "bar", ax = plt.gca())

for i, row in comparison_mean.iterrows():
    ax.annotate(row["track"][:25], (i, row["tiktok_views_song"]), ha = "center", va = "bottom", rotation = 75)
    
plt.xlabel("year")
plt.ylabel("tiktok views")
plt.xticks(rotation = 50)
plt.title("Comparison of most popular songs according to tiktok views to mean tiktok views for each year")

plt.tight_layout()
plt.show()
```

```python
top_3_songs = comparison_mean.sort_values(by="tiktok_views_song", ascending = False)[0:3]
top_3_songs
```

The song 'Monkeys Spinning Monkeys' released in 2011, 'Love you so' released in 2015, and 'Oh No' released in 2019 are top 3 most popular songs according to tiktok views. These 3 songs have large difference between their tiktok views and the mean for the year they were released.
